```python
# Concatenation 

import pandas as pd 

employees1 = pd.read_csv(r"C:\Users\AdmiN\Desktop\EK\B1_15May\Python\employees1.csv")
employees2 = pd.read_csv(r"C:\Users\AdmiN\Desktop\EK\B1_15May\Python\employees2.csv")
service = pd.read_csv(r"C:\Users\AdmiN\Desktop\EK\B1_15May\Python\service.csv")

print(employees1)
print(employees2)
print(service)
```

```python
import pandas as pd 
pd.concat([employees1,employees2])
```

```python
pd.concat([employees1,service], axis = 1)
```
