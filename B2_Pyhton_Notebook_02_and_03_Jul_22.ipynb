```python
## Line plot 

import numpy as np
import matplotlib.pyplot as plt 

data = np.array([1,2,4,7,9,12,14,17,19])

plt.plot(data, color = "g", linestyle = 'dotted')
```

```python
## Bar chart 

## Import 
import numpy as np 
import matplotlib.pyplot as plt 

## Data 
data = {"English": 89, 'Math': 67, 'Science': 90, 'History': 70}

Subject = list(data.keys())
Marks = list(data.values())

# Size of the plot 
fig = plt.figure(figsize=(8,2))

##Plot 
plt.bar(Subject,Marks, color = ['g','b','k','y'])

## Asthetics 
plt.title("This is my bar graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")

#plt.show()
```

```python
## Histogram 

import matplotlib.pyplot as plt 
import numpy as np

# Data 
data = np.random.normal(100,20,300)

# PLot 
plt.hist(data, bins = 10, orientation = 'vertical', color ='g')


plt.show()
```

```python
# Pie chart 
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
lab = 'English', 'Math', 'Science', 'History'
sizes = [85,89, 75, 90]

exp = (0, 0, 0.3, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

# Space
fig1, ax1 = plt.subplots()

# Chart
ax1.pie(sizes, explode=exp, labels=lab, autopct='%0.2f%%',
        shadow=True, startangle=0)

#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
